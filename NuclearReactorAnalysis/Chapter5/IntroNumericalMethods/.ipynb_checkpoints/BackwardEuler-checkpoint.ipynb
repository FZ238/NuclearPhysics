```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import symbols, solve
```

```python
#We'll be building from the forward Euler method, the only thing that changes is how the method works, all the other blocks are identical
#You can generalize this method for any DE of the form y'=f(x,y) with any initial condition and any step size.
def f(x,y):
 return 2*x

#You have to specify these numbers
x_init=0
y_init=0
x_end=0.8
h = 0.1

#Intermediary step to keep things tidy
Sol = [y_init]
n = round((x_end-x_init)/h)
y=symbols('y')

#The actual backward Euler method
#solve the implicit equation y_n+1-y_n-hf'(x_n+1,y_n+1), the solution is y_n+1 
for x in np.linspace(x_init, x_end, n, endpoint=False):
 expr = y-y_init-h*f(x+h,y)
 y_new = solve(expr, y)[0]  #in the above line we're subtracting y-y_init, the solve function returns a list though so in the next iteration you'll be comparing a list and a float.
 Sol.append(y_new)
 y_init=y_new

#Post-processing (Comment this out if it's not needed)
x = np.linspace(x_init, x_end, n+1)
plt.plot(x, Sol, color='b', label='Backward')
plt.legend()
plt.show()
```

```python
#As can be seen here, backward Euler will overshoot
x = np.linspace(x_init, x_end, n+1)
B = x**2
plt.plot(x, Sol, color='r', label='est')
plt.plot(x, B, color='g', label='real')
plt.legend()
plt.show()
```
