```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import symbols, solve
```

```python
#Illustrate the algorithm using a simple example 
#Given y'=2x with y(0)=0 and you're asked to find y on the interval [0, 0.8] using a step size of 0.1

#Name your variables
y_start=0
x_start=0
x_end=0.8
h=0.1 #step size

#Intermediary step of initializing your y-values and also to find the number of steps
A = [y_start]
n = round((x_end-x_start)/h) #This is a required input for np.linspace, do NOT use int as it'll always round down

#the actual implementation of the forward Euler method, y_n+1 = y_n + h * (y') where y' is based on the current iteration i.e. y'=f(x_n,y_n)
#Important note is that one may be tempted to use np.arange but this will lead to floating point errors 
#which will lead to a mismatch in size of your x- and y-values so better to just avoid it all together
#We don't include the endpoint otherwise we would calculate y(0.9)
for x in np.linspace(x_start, x_end, n,endpoint=False): 
 y=y_start+h*(2*x)
 A.append(y)
 y_start=y

#Quick comparison; Forward Euler will underestimate the actual values
#We do include 0.8 here since we have to plot y up until this value
x = np.linspace(x_start, x_end, n+1)
B = x**2
plt.plot(x, A, color='r', label='est')
plt.plot(x, B, color='g', label='real')
plt.legend()
plt.show()
```

```python
#You can generalize this method for any DE of the form y'=f(x,y) with any initial condition and any step size.
def f(x,y):
 return 2*x

#You have to specify these numbers
y_init=0
x_init=0
x_end=0.8
h = 0.1 #step size

#Intermediary step to keep things tidy
Sol = [y_init]
n = round((x_end-x_init)/h)

#The actual forward Euler method
for x in np.linspace(x_init, x_end, n, endpoint=False):
 y=y_init+h*f(x,y_init)
 Sol.append(y)
 y_init=y

#Post-processing (Comment this out if it's not needed)
x = np.linspace(x_init, x_end, n+1)
plt.plot(x, Sol, color='b', label='Forward')
plt.legend()
plt.show()
```
