```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import symbols, solve
```

```python
#You can generalize this method for any DE of the form y'=f(x,y) with any initial condition and any step size.
def f(x,y):
 return y*(1-y)

#You have to specify these numbers
x_init=0
y_init=2
x_end=2
h = 0.1

#Intermediary step to keep things tidy
Sol = [y_init]
n = round((x_end-x_init)/h)

#Runge Kutta implementation
for x in np.linspace(x_init, x_end, n, endpoint=False):
 k1=f(x,y_init)
 k2=f(x+h/2,y_init+k1/2)
 k3=f(x+h/2,y_init+k2/2)
 k4=f(x+h,y_init+k3)

 y_new=y_init+h/6*(k1+2*k2+2*k3+k4)
 Sol.append(y_new)
 y_init=y_new

#Post-processing (Comment this out if it's not needed)
x = np.linspace(x_init, x_end, n+1)
plt.plot(x, Sol, color='b', label='Backward')
plt.legend()
plt.show()
```

```python
#Compare this to the true solution i.e. y=2/(2-e^(-x))
x = np.linspace(x_init, x_end, n+1)
B = 2/(2-np.exp(-x))
plt.plot(x, Sol, color='b', label='RK4')
plt.plot(x, Sol, color='g', label='True')
plt.legend()
plt.show()
```
