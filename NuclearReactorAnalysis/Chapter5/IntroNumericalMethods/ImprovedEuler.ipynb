```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import symbols, solve
```

```python
#We'll be building from both the forward and backward Euler method. The improved Euler method (or trapezoidal method) is a continuitation from these two
#and will serve as a foundation for the Runge-Kutta method (known as RK4)
#You can generalize this method for any DE of the form y'=f(x,y) with any initial condition and any step size.
def f(x,y):
 return 2*x

#You have to specify these numbers
x_init=0
y_init=0
x_end=0.8
h = 0.1

#Intermediary step to keep things tidy
Sol = [y_init]
n = round((x_end-x_init)/h)

#The improved Euler method
for x in np.linspace(x_init, x_end, n, endpoint=False):
 y_pred=y_init+h*f(x,y_init)

 y_new=y_init+h/2*(f(x,y_init)+f(x+h,y_pred))
 Sol.append(y_new)
 y_init=y_new

#Post-processing (Comment this out if it's not needed)
x = np.linspace(x_init, x_end, n+1)
plt.plot(x, Sol, color='b', label='Improved Euler')
plt.legend()
plt.show()
```

```python
#As can be seen here, in the case of y'=2x we average out the overshooting from the backward Euler with the undershooting from the forward Euler perfectly
#In general these effects don't cancel each other out but u do end up with a solution that's closer to the true solution compared to the individual methods 
x = np.linspace(x_init, x_end, n+1)
B = x**2
plt.plot(x, Sol, color='r', label='est')
plt.plot(x, B, color='g', label='real')
plt.legend()
plt.show()
```
